## Normal K-fold by 5

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import KFold, cross_val_predict
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Load the data
inflation_data_path = 'expanded2_usa_inflation_rate.csv'
bitcoin_data_path = 'data_btc.csv'

inflation_data = pd.read_csv(inflation_data_path)
bitcoin_data = pd.read_csv(bitcoin_data_path)

# Preprocessing
inflation_data['date'] = pd.to_datetime(inflation_data['date'], dayfirst=True)
bitcoin_data['date'] = pd.to_datetime(bitcoin_data['date'], dayfirst=True)

# Merging the data on the 'date' column
merged_data = pd.merge(bitcoin_data, inflation_data, on='date', how='inner')
merged_data_final = merged_data[['date', 'price', 'total_volumes', 'Monthly Inflation Rate', 'Annual Inflation Rate']]

# Handling NaN values by filling them with the mean of their respective columns
merged_data_final_filled = merged_data_final.fillna(merged_data_final.mean(numeric_only=True))

# Features and target variable
X = merged_data_final_filled[['Monthly Inflation Rate', 'total_volumes']]
y = merged_data_final_filled['price']

# K-Fold setup
kf = KFold(n_splits=5, random_state=60, shuffle=True)
model = LinearRegression()

# Perform cross-validation and store predictions for each fold
predictions = cross_val_predict(model, X, y, cv=kf)

# Convert dates to datetime objects
dates = merged_data_final_filled['date']
dates = pd.to_datetime(dates)

# Merging the data on the 'date' column
merged_data = pd.merge(bitcoin_data, inflation_data, on='date', how='inner')
merged_data_final = merged_data[['price', 'Monthly Inflation Rate', 'total_volumes']]

# Plotting correlation matrix heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = merged_data_final.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 12})
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Initialize lists to store performance metrics for each fold
r2_scores = []
rmse_scores = []

# Iterate over each fold
for train_index, test_index in kf.split(X):
    # Split data into train and test sets
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Fit the model on the training data
    model.fit(X_train, y_train)
    
    # Make predictions on the test data
    y_pred = model.predict(X_test)
    
    # Calculate R-squared score
    r2 = r2_score(y_test, y_pred)
    
    # Calculate RMSE
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    
    # Append scores to the respective lists
    r2_scores.append(r2)
    rmse_scores.append(rmse)

# Print the R-squared scores and RMSE for each fold
for i in range(len(r2_scores)):
    print(f"Fold {i+1}: R-squared = {r2_scores[i]:.4f}, RMSE = {rmse_scores[i]:.4f}")

# Calculate average R-squared score and RMSE across all folds
avg_r2 = np.mean(r2_scores)
avg_rmse = np.mean(rmse_scores)
print("\nAverage R-squared across all folds:", avg_r2)
print("Average RMSE across all folds:", avg_rmse)
```

```python
# Scatter plot for each fold's predictions
plt.figure(figsize=(5, 20))

for i, (train_index, test_index) in enumerate(kf.split(X), 1):
    plt.subplot(6, 1, i)
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    model.fit(X_train, y_train)
    fold_pred = model.predict(X_test)
    plt.scatter(y_test, fold_pred, label=f'Fold {i}', color='blue', alpha=0.5)
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')  # Identity line
    plt.title(f'Fold {i}')
    plt.xlabel('Actual Price')
    plt.ylabel('Predicted Price')
    plt.legend()

# Scatter plot for average predictions
plt.subplot(6, 1, 6)
plt.scatter(y, predictions, label='Average Predicted', color='green', alpha=0.5)
plt.plot([min(y), max(y)], [min(y), max(y)], color='red', linestyle='--')  # Identity line
plt.title('Average Predicted')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Plotting predicted prices for each fold
plt.figure(figsize=(10, 22))

for i, (train_index, test_index) in enumerate(kf.split(X), 1):
    plt.subplot(6, 1, i)
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    model.fit(X_train, y_train)
    fold_pred = model.predict(X_test)
    plt.scatter(dates[test_index], y.iloc[test_index], label='Actual', color='blue', alpha=0.7)
    plt.plot(dates[test_index], fold_pred, linestyle='--', color='red', label='Predicted')
    plt.title(f'Fold {i}')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()

# Plotting average predicted values
plt.subplot(6, 1, 6)
plt.plot(dates, y, label='Actual', color='blue')
plt.plot(dates, predictions, label='Average Predicted', color='red', linestyle='--')
plt.title('Average Predicted')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

plt.tight_layout()
plt.show()
```

## Specified kbins test 1

using manual trainig range and testing on the same range

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Load the data
inflation_data_path = 'expanded2_usa_inflation_rate.csv'
bitcoin_data_path = 'data_btc.csv'

inflation_data = pd.read_csv(inflation_data_path)
bitcoin_data = pd.read_csv(bitcoin_data_path)

# Preprocessing
inflation_data['date'] = pd.to_datetime(inflation_data['date'], dayfirst=True)
bitcoin_data['date'] = pd.to_datetime(bitcoin_data['date'], dayfirst=True)

# Merging the data
merged_data = pd.merge(bitcoin_data, inflation_data, on='date', how='inner')
merged_data_final = merged_data[['date', 'price', 'total_volumes', 'Monthly Inflation Rate', 'Annual Inflation Rate']]

# Handling NaN values by filling them with the mean of their respective columns
merged_data_final_filled = merged_data_final.fillna(merged_data_final.mean(numeric_only=True))

# Features and target variable
X_filled = merged_data_final_filled[['Monthly Inflation Rate', 'total_volumes']]
y_filled = merged_data_final_filled['price']

# Define custom date ranges for training and testing
fold_date_ranges = [
    ("2014-01-01", "2016-01-01", "2018-01-01"),
    ("2016-01-01", "2018-01-01", "2020-01-01"),
    ("2017-01-01", "2019-01-01", "2021-01-01"),
    ("2018-01-01", "2020-01-01", "2022-01-01"),
    ("2019-01-01", "2021-01-01", "2023-01-01")
]

mse_scores_custom_folds = []

# Iterate over each custom fold
for start_date, train_end_date, test_end_date in fold_date_ranges:
    # Split the dataset based on the date range
    train_mask = (merged_data_final_filled['date'] >= start_date) & (merged_data_final_filled['date'] < train_end_date)
    test_mask = (merged_data_final_filled['date'] >= train_end_date) & (merged_data_final_filled['date'] < test_end_date)
    
    X_train, y_train = X_filled[train_mask], y_filled[train_mask]
    X_test, y_test = X_filled[test_mask], y_filled[test_mask]
    
    # Train the Linear Regression model
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # Make predictions and adjust negative predictions to zero
    y_pred = model.predict(X_test)
    y_pred[y_pred < 0] = 0
    
    # Calculate and append the MSE for the current fold
    mse = mean_squared_error(y_test, y_pred)
    mse_scores_custom_folds.append(mse)

# Output the MSE scores for each fold and their average
print(f"MSE scores for each custom fold: {mse_scores_custom_folds}")
print(f"Average MSE across all custom folds: {np.mean(mse_scores_custom_folds)}")
```

```python
import matplotlib.pyplot as plt

# Additional setup for plotting
plt.figure(figsize=(15, 12))

for i, (start_date, train_end_date, test_end_date) in enumerate(fold_date_ranges, 1):
    train_mask = (merged_data_final_filled['date'] >= start_date) & (merged_data_final_filled['date'] < train_end_date)
    test_mask = (merged_data_final_filled['date'] >= train_end_date) & (merged_data_final_filled['date'] < test_end_date)
    
    X_train, y_train = X_filled[train_mask], y_filled[train_mask]
    X_test, y_test = X_filled[test_mask], y_filled[test_mask]
    
    # Train the Linear Regression model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    y_pred[y_pred < 0] = 0
    
    # Plotting
    plt.subplot(3, 2, i)
    plt.plot(merged_data_final_filled['date'][test_mask], y_test, label='Actual Price', color='red')
    plt.plot(merged_data_final_filled['date'][test_mask], y_pred, label='Predicted Price', linestyle='--', color='blue')
    plt.title(f'Fold {i}: {train_end_date} to {test_end_date}')
    plt.xlabel('Date')
    plt.ylabel('Bitcoin Price')
    plt.xticks(rotation=45)
    plt.legend()

plt.tight_layout()
plt.show()
```

## Manual Kbins test 2

using specific trainig range and testing on the rest 

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt

# Load and preprocess your data
inflation_data_path = 'expanded2_usa_inflation_rate.csv'  # Update this path
bitcoin_data_path = 'data_btc.csv'  # Update this path

inflation_data = pd.read_csv(inflation_data_path)
bitcoin_data = pd.read_csv(bitcoin_data_path)

inflation_data['date'] = pd.to_datetime(inflation_data['date'], dayfirst=True)
bitcoin_data['date'] = pd.to_datetime(bitcoin_data['date'], dayfirst=True)

merged_data = pd.merge(bitcoin_data, inflation_data, on='date', how='inner')
merged_data_final = merged_data[['date', 'price', 'total_volumes', 'Monthly Inflation Rate', 'Annual Inflation Rate']]
merged_data_final_filled = merged_data_final.fillna(merged_data_final.mean(numeric_only=True))

# Define training and testing periods
training_periods = [
    (datetime(2014, 1, 1), datetime(2016, 12, 31)),
    (datetime(2016, 1, 1), datetime(2018, 12, 31)),
    (datetime(2017, 1, 1), datetime(2019, 12, 31)),
    (datetime(2018, 1, 1), datetime(2020, 12, 31)),
    (datetime(2019, 1, 1), datetime(2021, 12, 31)),
]

# Initialize model
model = LinearRegression()

# Plotting setup
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14, 18))
axes = axes.flatten()  # Flatten for easy iteration

for i, (start_date, end_date) in enumerate(training_periods):
    train_data = merged_data_final_filled[(merged_data_final_filled['date'] >= start_date) & (merged_data_final_filled['date'] <= end_date)]
    test_data = merged_data_final_filled[(merged_data_final_filled['date'] < start_date) | (merged_data_final_filled['date'] > end_date)]
    
    # Train the model
    X_train = train_data[['Monthly Inflation Rate', 'total_volumes']]
    y_train = train_data['price']
    model.fit(X_train, y_train)
    
    # Predictions for plotting
    combined_data = pd.concat([train_data, test_data]).sort_values(by='date')
    X_combined = combined_data[['Monthly Inflation Rate', 'total_volumes']]
    y_combined = combined_data['price']
    y_pred_combined = model.predict(X_combined)
    
    # Plot actual vs predicted prices
    ax = axes[i]
    ax.plot(combined_data['date'], y_combined, label='Actual Price', color='red')
    ax.plot(combined_data['date'], y_pred_combined, label='Predicted Price', linestyle='--', color='blue')
    ax.set_title(f'Train: {start_date.year}-{end_date.year}, Test: Rest')
    ax.set_xlabel('Date')
    ax.set_ylabel('Bitcoin Price (USD)')
    ax.legend()

plt.tight_layout()
plt.show()
```
